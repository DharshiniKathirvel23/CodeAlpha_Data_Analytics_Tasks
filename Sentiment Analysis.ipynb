```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Review": [
        "This product is amazing",
        "Very bad experience",
        "The phone is good",
        "Waste of money",
        "I love this product",
        "Average quality",
        "Bad performance",
        "Not worth the price"
    ]
}

df = pd.DataFrame(data)

df
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["Review"].apply(get_sentiment)

df
```

```python
df["Sentiment"].value_counts()
```

```python
df["Sentiment"].value_counts().plot(kind="bar")

plt.title("Sentiment Analysis Results")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")

plt.show()
```

Conclusion

This project performs sentiment analysis on product reviews using Python and TextBlob.

The reviews were classified into Positive, Negative, and Neutral categories. 
Visualization shows the distribution of sentiments in the dataset.

Sentiment analysis helps businesses understand customer opinions and improve products.
